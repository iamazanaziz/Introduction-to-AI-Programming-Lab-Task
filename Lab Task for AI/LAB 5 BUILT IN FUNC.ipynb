## task1 : enumerate () adds numbers (index) to items in a list or other collection.  
i.	A list contains voltage readings from 5 measurement points. Use enumerate () to display the reading   number with each voltage.  
ii.	A student records resistance values of resistors in a lab. Use enumerate () to print the resistor  number and its value.  


```python
# Task 1:enumerate() 
voltages = [200, 230, 225, 235, 240]
print("Voltage Readings:")
for i,v in enumerate(voltages):
    print( i + 1, ":", v, "V")

resistances = [100, 220, 330, 470, 560]
print("\nResistor Values:")
for index, value in enumerate(resistances):
    print(index + 1, ":", value, "Ohms")
```

## task: 2: zip () combines two or more lists element by element.  
i.	Two lists, one list has current readings, and another has voltage readings. Use zip() to print the pair of  current and voltage.  
ii.	Combine a list of component names with their ratings (resistor with ohms, capacitor with µF).  


```python
# Task 2: zip() 
currents = [5, 10, 15, 20]
voltages = [110, 220, 330, 440]
pairs=list(zip(currents,voltages))
print("Current and Voltage Pairs:")
print(pairs)

components = ["Resistor", "Capacitor", "Inductor"]
ratings = ["100 ohm", "10 µF", "1 H"]
print("\nComponents and Ratings:")
for name, rating in zip(components, ratings):
    print(name, ":", rating)
```

## task 3: map() applies a function to every item in a list.  
Note: lambda is a short way to write a function in one line. Example: lambda x: x*2 means a function that doubles x.  
i.	Convert a list of temperatures from Celsius to Fahrenheit using the formula F = (C × 9/5) + 32.  
ii.	Convert a list of voltages in volts to kilovolts using V_kV = V / 1000.


```python
# Task 3: map()
temp_c = [0, 25, 50, 100]
def c_to_f(x):
    return (x * 9/5) + 32
temp_f = list(map(c_to_f, temp_c))
print("Temperatures in Fahrenheit:", temp_f)
voltages_v = [1000, 2000, 3000]
def v_to_kv(v):
    return v / 1000
voltages_kv = list(map(v_to_kv, voltages_v))
print("Voltages in kV:", voltages_kv)
```

## task 4:filter() selects only the items that match a condition.  
i.	A list contains current readings. Use filter() to keep only readings greater than 10 A.  
ii.	A capacitor bank has values [10, 20, 50, 100]. Use filter() to select capacitors above 50 µF.


```python
# Task 4: filter()
currents = [5, 12, 8, 15, 3]
def greater_than_10(x):
    return x > 10
filtered = list(filter(greater_than_10, currents))
print("Currents greater than 10A:", filtered)
capacitors = [10, 20, 50, 100]
def above_50(x):
    return x > 50
filtered_caps = list(filter(above_50, capacitors))
print("Capacitors above 50 µF:", filtered_caps)
```

## task 18: divmod() returns the quotient and remainder of the division.  
i.	87 optical fiber cables are drawn to provide high-speed internet for a university campus. The   duct can contain 10 cables at most. Divide the total wires into bundles and find the number of   completed filed ducts and how many more cables can be placed in the final duct.
ii.	Each group brings 42 LEDS for the DLD project. Divide the total number of LEDs into groups of 12, and find the groups and extra LEDs.


```python
# Task 18: divmod()
cables = 87
duct_capacity = 10
result = divmod(cables, duct_capacity)
print("Cables division result:", result)
leds = 42
group_size = 12
result2 = divmod(leds, group_size)
print("LED groups and extra LEDs:", result2)
```
