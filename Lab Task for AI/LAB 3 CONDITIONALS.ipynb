## Task:1 Voltage Level Classification  
i.	A text file contains circuit voltage readings (one per line).  
ii.	Task: Read the file and classify each voltage as:  
a.	Low (< 110V)  
b.	Medium (110V–220V)  
c.	High (> 220V)  
iii.	Save the classification results into a new file.


```python
with open(r"voltages5.txt","r") as f:
    voltages = f.readlines()

results = []
for v in voltages:
    voltages = int(v.strip())
    if voltages < 110:
        results.append(f"{voltages}v low")
    elif 110<=voltages<=220:
        results.append(f"{voltages}v medium")
    else :
        results.append(f"{voltages}v high")
print(results)
with open ("classified.txt","w") as file:
    for line in results:
        file.write(f"{line} \n")
    print("result saved in classificatied.txt")
```

## Task:2 Pass/Fail Test Report (Engineering Labs)  
i.	A file stores students’ lab test scores.  
ii.	Task: Read scores from file and apply conditions:  
a.	Pass if ≥ 50.  
b.	Fail if < 50.


```python

with open("score.txt","r") as f:
    scores = f.readlines()

results = []
for s in scores:
    marks = int(s.strip())
    if marks >= 50:
        results.append(f"{s.strip()} Pass")
    else:
        results.append("Fail")

with open("results.txt", "w") as f:
    for r in results:
        f.write(r + "\n")
    print(results)
```

## Task:3 Current Rating Check (Electrical Machines)  
i.	A file has current values (in Amperes).  
ii.	Task: Check for each line:  
a.	If current < 5A → Safe  
b.	If 5A ≤ current < 10A → Warning  
c.	If ≥ 10A → Overload


```python
with open(r"current.txt", "r") as f:
    currents = f.readlines()

results = []
for c in currents:
    current = float(c.strip())
    if current < 5:
        results.append("Safe")
    elif current < 10:
        results.append("Warning")
    else:
        results.append("Overload")

with open("current_results.txt", "w") as f:
    for r in results:
        f.write(r + "\n")
    print(results)
```

## Task:4 File Extension Validator (General Engineering Reports)
i.	A directory has multiple engineering reports.  
ii.	Task: Write a program to scan all files.  
a.	If file ends with .pdf → Valid Report  
b.	If file ends with .docx → Needs Conversion  
c.	Otherwise → Invalid Format  
iii.	Count and display the numbers of each type.


```python
import os

valid = needs_conv = invalid = 0 # veriables to count valid files etc

for filename in os.listdir(r"D:\INTAIPRO"): # making list of file names in given folder path
    if filename.endswith(".pdf"): #using endswith built-in funtion to cheak types of files.
        print(filename, "→ valid")
        valid += 1
    elif filename.endswith(".docx"):
        print(filename, "→ Needs Conversion")
        needs_conv += 1
    else:
        print(filename, "→ Invalid Format")
        invalid += 1

print("\nSummary:") 
print("Valid Reports:", valid)
print("Needs Conversion:", needs_conv)
print("Invalid Format:", invalid)
```
