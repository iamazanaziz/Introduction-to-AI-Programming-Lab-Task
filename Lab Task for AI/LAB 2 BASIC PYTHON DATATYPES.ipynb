## 2. File Handling : Integer , Number and String

```python
import csv

def add_two(value):
    try:
        return int(value) + 2
    except ValueError:
        return value + "2"

with open('numbers.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        updated_row = [add_two(item) for item in row]
        print(updated_row)
```

## 3. TASK  
   you are given a CSV file named integer.csv that contains one number per row (some 
are integers, and some are decimal numbers).

```python
import csv

with open('integers.csv', newline='') as file:
    reader = csv.reader(file)
    numbers = [float(row[0]) for row in reader]

integer_count = sum(num.is_integer() for num in numbers)

print("Total count of integers:", integer_count)
```

## 5. Task  
Represent Impedance Using Complex Numbers 

```python
import cmath
import math
# Circuit parameters
R = 50 # ohms
L = 0.1 # H
C = 100e-6 # F
f = 60 # Hz
w = 2 * math.pi * f
# Impedances
Z_R = complex(R, 0)
Z_L = complex(0, w * L)
Z_C = complex(0, -1 / (w * C))
print("Resistor Impedance:", Z_R)
print("Inductor Impedance:", Z_L)
print("Capacitor Impedance:",Z_C)
```

## 6. Task 
 Find Total Impedance

```python
import cmath

V = complex(120, 0)   

Z = complex(30, 40)   

I = V / Z

print("Current phasor (I):", I)
print("Magnitude of Current |I| =", abs(I))
print("Phase angle of Current (radians) =", cmath.phase(I))
print("Phase angle of Current (degrees) =", cmath.phase(I) * 180 / 3.14159)
```

## 7. Task 
      Compute Circuit Current 

```python
import cmath
import math

V = complex(120, 0)      
Z = complex(30,40)       

I = V / Z

magnitude = abs(I)
phase = math.degrees(cmath.phase(I))

print("Current magnitude =", magnitude, "A")
print("Current phase =", phase, "degrees")
```

## 8. Task: Run and understand  
###     a. Ohm’s Law 

```python
V = 230 # volts
I = 5 # amps
R = V / I # ohms
# Using f-string
print(f"The voltage is {V} V, the current is {I} A, so the resistance is {R:.2f} Ω.")
# Using .format()
print("The voltage is {} V, the current is {} A, so the resistance is {:.2f} Ω.".format(V, I, R))
```

### b. Efficiency Calculation 

```python
P_out = 950 # watts
P_in = 1000 # watts
eff = (P_out / P_in) * 100
print(f"The machine efficiency is {eff:.1f}% when output is {P_out} W and input is {P_in} W.")
print("The machine efficiency is {:.1f}% when output is {} W and input is {} W.".format(eff, P_out, P_in))
```

### c. AC Power 

```python
P = 2000     # watts 
V = 220      # volts 
I = P / V    # amps
print(f"The real power is {P} W at {V} V supply, drawing {I:.2f} A current.") 
print("The real power is {0} W at {1} V supply, drawing {2:.2f} A current.".format(P, V, I))
```

### d. RLC Circuit Impedance 

```python
R = 50
X_L = 30
X_C = 20
Z = complex(R, X_L - X_C)
print(f"Impedance of the circuit is {Z.real} + j{Z.imag} Ω.")
print("Impedance of the circuit is {0} + j{1} Ω.".format(Z.real,Z.imag))
```

## 9. Task  
Write a Python program that: 

Reads a given directory path. 

Checks all files in that directory. 

Counts how many files are PDF files (ending with .pdf). 

Prints the total number of PDF files. 

```python
import os

# Enter directory path here
directory_path = "D:\AI ASSIG"

pdf_count = 0
for file in os.listdir(directory_path):
    if file.lower().endswith(".pdf"):
        pdf_count += 1
print("Total number of PDF files:", pdf_count)
```
