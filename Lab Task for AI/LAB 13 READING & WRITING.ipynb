### Load Demand Data Recording
An electrical feederâ€™s hourly load demand (kW) needs to be recorded for monitoring.
Tasks:
Create a CSV file to store hourly load data
Write load values with time stamps
Organize each record on a separate line


```python
import csv

file = open("load_demand.csv", "w", newline="")
writer = csv.writer(file)

writer.writerow(["Time", "Load (kW)"])
writer.writerow(["01:00", 120])
writer.writerow(["02:00", 135])
writer.writerow(["03:00", 128])

file.close()
```

### Solar Energy Generation Logging
A solar PV system records daily energy generation (kWh).
Tasks:

### Task: Organize and Process CSV Data
Read the CSV file line by line and:
Count the total number of recorded days
Display each record with a serial number
The header must not be counted as a data record.


```python
import csv

file = open("energy_data.csv", "r")
reader = csv.reader(file)

count = 0
header = next(reader)

for row in reader:
    count += 1
    print(count, row)

file.close()
print("Total recorded days:", count)
```

### Task: Write Organized Data to a New CSV File
Create a new file named organized_energy_data.csv.
Write the processed and formatted energy records into this file in CSV format.
The original file must remain unchanged.


```python
import csv

input_file = open("energy_data.csv", "r")
reader = csv.reader(input_file)

output_file = open("organized_energy_data.csv", "w", newline="")
writer = csv.writer(output_file)

header = next(reader)
writer.writerow(["S.No"] + header)

serial = 1
for row in reader:
    writer.writerow([serial] + row)
    serial += 1

input_file.close()
output_file.close()

print("Organized energy data written to new file.")
```
