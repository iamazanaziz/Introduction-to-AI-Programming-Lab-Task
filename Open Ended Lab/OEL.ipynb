### Take the CityLearn dataset of the 2022 competition. Load the data of building_2 and perform the following.

1) LOAD DATASET AND STORE IN LIST

```python
file = open("Building_2.csv", "r")

lines = file.readlines()
file.close()

data = []

for i in range(1, len(lines)):
    row = lines[i].strip().split(",")
    data.append(row)
```

2. ANNUAL ENERGY CONSUMPTION

```python
annual_consumption = 0

for row in data:
    equipment = float(row[7])
    dhw = float(row[8])
    cooling = float(row[9])
    heating = float(row[10])

    total = equipment + dhw + cooling + heating
    annual_consumption += total
print("\nANNUAL ENERGY CONSUMPTION SUMMARY")
print("-" * 50)


scale = 35  # width of bar

bar_length = int((annual_consumption / annual_consumption) * scale)
bar = "█" * bar_length
print(f"Annual| {bar:<35} {annual_consumption:.2f}")
```

3. MONTHLY ENERGY CONSUMPTION

```python
monthly_consumption = {}

for row in data:
    month = int(float(row[0])) 

    total = float(row[7]) + float(row[8]) + float(row[9]) + float(row[10])

    if month not in monthly_consumption:
        monthly_consumption[month] = 0

    monthly_consumption[month] += total

print("\nMONTHLY ENERGY CONSUMPTION (Building 2)")
print("-" * 50)

max_value = max(monthly_consumption.values())
scale = 40  # width of bar

for m in sorted(monthly_consumption):
    value = monthly_consumption[m]
    bar_length = int((value / max_value) * scale)
    bar = "█" * bar_length

    print(f"Month {m:2} | {bar:<40} {value:.2f}")
```

4. DAILY ENERGY CONSUMPTION 

```python
daily_consumption = {}

day = 1
hour_count = 0
daily_consumption[day] = 0

for row in data:
    total = float(row[7]) + float(row[8]) + float(row[9]) + float(row[10])
    daily_consumption[day] += total
    hour_count += 1

    if hour_count == 24:
        day += 1
        daily_consumption[day] = 0
        hour_count = 0

print("\nDAILY ENERGY CONSUMPTION (First 10 Days)")
print("-" * 50)

max_day = max(daily_consumption[d] for d in range(1, 11))
scale = 40

for d in range(1, 11):
    value = daily_consumption[d]
    bar_len = int((value / max_day) * scale)
    bar = "█" * bar_len
    print(f"Day {d:2} | {bar:<40} {value:.2f}")

```

5. SEASONAL ENERGY CONSUMPTION

```python
season_consumption = {
    "Winter": 0,
    "Spring": 0,
    "Summer": 0,
    "Autumn": 0
}

for row in data:
    month = int(float(row[0]))
    total = float(row[7]) + float(row[8]) + float(row[9]) + float(row[10])

    if month in [12, 1, 2]: 
            #Winter: Dec–Feb
        season_consumption["Winter"] += total
    elif month in [3, 4, 5]:    
            #Spring: Mar–May
        season_consumption["Spring"] += total
    elif month in [6, 7, 8]:    
            #Summer: Jun–Aug
        season_consumption["Summer"] += total
    else:                       
            #Autumn: Sep–Nov
        season_consumption["Autumn"] += total

print("\nSEASONAL ENERGY CONSUMPTION")
print("-" * 50)

max_season = max(season_consumption.values())

for season in season_consumption:
    value = season_consumption[season]
    bar_len = int((value / max_season) * 40)
    bar = "█" * bar_len
    print(f"{season:<7} | {bar:<40} {value:.2f}")
```

6. WEEKDAY VS WEEKEND CONSUMPTION

```python
weekday_consumption = 0
weekend_consumption = 0

day_number = 1
hour_count = 0

for row in data:
    total = float(row[7]) + float(row[8]) + float(row[9]) + float(row[10])

    # Weekday: Mon(1)–Fri(5), Weekend: Sat(6), Sun(7)
    if day_number % 7 == 6 or day_number % 7 == 0:
        weekend_consumption += total
    else:
        weekday_consumption += total

    hour_count += 1

    if hour_count == 24:
        day_number += 1
        hour_count = 0

print("\nWEEKDAY vs WEEKEND ENERGY CONSUMPTION")
print("-" * 50)

total = weekday_consumption + weekend_consumption
weekday_bar = int((weekday_consumption / total) * 40)
weekend_bar = int((weekend_consumption / total) * 40)

print(f"Weekday | {'█' * weekday_bar:<40} {weekday_consumption:.2f}")
print(f"Weekend | {'█' * weekend_bar:<40} {weekend_consumption:.2f}")
```
